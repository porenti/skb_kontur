# Отчистка данных
Приведем заголовки к единому стилю, исключив пунктуацию, шумовые слова и изменим регистр.
Из-за этого введем переменную с очищенными данными.

**1. Приведение к общему регистру**


```python
clear_data = dirt_data['title'].apply(lambda x: x.lower())
clear_data
```

**2. Удаление лишних символов**

```python
def punctuation_remove(text):
    '''Удаляет лишние символы'''
    chars = string.punctuation
    chars += '«»' #не входят в punctuation
    all_list = [char for char in text if char not in chars]
    clean_str = ''.join(all_list)
    
    return clean_str


clear_data = clear_data.apply(punctuation_remove)
clear_data
```

**3. Удаление шумовых слов (стоп-слова)**

***Список стоп-слов***

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
stop = stopwords.words("russian")
print("  ".join(stop))
```

```python
clear_data = clear_data.apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))
clear_data
```

**4. Отчистка лишних пробелов**

```python
from nltk.tokenize import TreebankWordTokenizer
clear_data = clear_data.apply(lambda x: TreebankWordTokenizer().tokenize(x))
clear_data
```

**5. Привение к нормальной форме**

```python
morph = pymorphy2.MorphAnalyzer(lang='ru')

clear_data = clear_data.apply(lambda x: [morph.parse(str(item))[0][2] for item in x])
clear_data
```

**6. Возврат значений ложных и правдивых заголовков**

Введем переменную для итогового результата finish_clear_data

```python
finish_clear_data = pd.concat([clear_data, dirt_data["is_fake"]], axis=1)
finish_clear_data
```

**7. Возврат строки на место списков**

```python
finish_clear_data['title'] = finish_clear_data['title'].apply(lambda x: ' '.join(x))
finish_clear_data
```

***Скомпонуем отчистку в одну функцию***

```python
def clean_data_frame(_data_frame):  #передаем датафрейм с заголовками
    _data_frame['title'] = _data_frame['title'].apply(lambda x: x.lower())
    _data_frame['title'] = _data_frame['title'].apply(punctuation_remove)
    _data_frame['title'] = _data_frame['title'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))
    _data_frame['title'] = _data_frame['title'].apply(lambda x: TreebankWordTokenizer().tokenize(x))
    _data_frame['title'] = _data_frame['title'].apply(lambda x: [morph.parse(str(item))[0][2] for item in x])
    _data_frame['title'] = _data_frame['title'].apply(lambda x: ' '.join(x))
    return _data_frame
```

**Чаще всего встречающиейся слова в фейковых новостях**

```python
token_space = nltk.tokenize.WhitespaceTokenizer()

def counter(text, column_text, quantity):
    all_words = ' '.join([text for text in text[column_text]])
    token_phrase = token_space.tokenize(all_words)
    frequency = nltk.FreqDist(token_phrase)
    df_frequency = pd.DataFrame({"Word": list(frequency.keys()),
                                   "Frequency": list(frequency.values())})
    df_frequency = df_frequency.nlargest(columns = "Frequency", n = quantity)
    plt.figure(figsize=(12,8))
    ax = sns.barplot(data = df_frequency, x = "Word", y = "Frequency", color = 'blue')
    ax.set(ylabel = "Count")
    plt.xticks(rotation='vertical')
    plt.show()
    
counter(finish_clear_data[finish_clear_data["is_fake"] == 1], "title", 20)
```

# Моделирование и подгонка моделей

**Для моделирования будем использовать 3 модели:**

**1. Логистическая регрессия (префикс "l_")**

**2. Дерево решений (префикс "d_")**

**3. Random forest (случайный лес)  (префикс "rf_")**

---
**Точность каждой модели проверим двумя способами:**
 1. **Обучаем на 80% данных и проверяем на 20% оставшихся**
 2. **Обучаем на всех данных и проверяем на 20% случайных (префикс full)**
---

**Ввод новой переменной, чтобы не испорить старую и не возвращаться к прошлому коду**

```python
_finish_clear_data = finish_clear_data
_finish_clear_data
```

## Логистическая регрессия

```python
X_train,X_test,y_train,y_test = train_test_split(_finish_clear_data['title'], _finish_clear_data['is_fake'], test_size=0.2, random_state=9)


pipe = Pipeline([('vect', CountVectorizer()),
                 ('tfidf', TfidfTransformer()),
                 ('model', LogisticRegression())])

l_model = pipe.fit(X_train, y_train)
l_prediction = l_model.predict(X_test)
print("Точность на обучении 80% данных: {}%".format(round(metrics.accuracy_score(y_test, l_prediction)*100,2)))

l_full_model = pipe.fit(_finish_clear_data['title'], _finish_clear_data['is_fake'])
l_full_prediction = l_full_model.predict(X_test)
print("Точность на обучении всеми данными: {}%".format(round(metrics.accuracy_score(y_test, l_full_prediction)*100,2)))
```

### Функция для пострания матрицы ошибок
***Матрица для 80% логистической регрессии***

```python
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")
plt.tight_layout()
plt.ylabel('True label')
plt.xlabel('Predicted label')

cm = metrics.confusion_matrix(y_test, l_prediction)
plot_confusion_matrix(cm, classes=['Fake', 'Real'])
```

***Матрица для 100% логистической регрессии***

```python
cm = metrics.confusion_matrix(y_test, l_full_prediction)
plot_confusion_matrix(cm, classes=['Fake', 'Real'])
```

## Дерево решений


```python
pipe = Pipeline([('vect', CountVectorizer()),
                 ('tfidf', TfidfTransformer()),
                 ('model', DecisionTreeClassifier(criterion= 'entropy', 
                                           splitter='random', 
                                           random_state=6,
                                           max_leaf_nodes= 1000))])

d_model = pipe.fit(X_train, y_train)
d_prediction = d_model.predict(X_test)
print("Точность на обучении 80% данных: {}%".format(round(metrics.accuracy_score(y_test, d_prediction)*100,2)))

d_full_model = pipe.fit(_finish_clear_data['title'], _finish_clear_data['is_fake'])
d_full_prediction = d_full_model.predict(X_test)
print("Точность на обучении всеми данными: {}%".format(round(metrics.accuracy_score(y_test, d_full_prediction)*100,2)))
```

***Матрица для 80% обучения дерева решений***

```python
cm = metrics.confusion_matrix(y_test, d_prediction)
plot_confusion_matrix(cm, classes=['Fake', 'Real'])
```

***Матрица для полностью обученного дерева решений***

```python
cm = metrics.confusion_matrix(y_test, d_full_prediction)
plot_confusion_matrix(cm, classes=['Fake', 'Real'])
```

## Модель случайного леса:

```python
pipe = Pipeline([('vect', CountVectorizer()),
                 ('tfidf', TfidfTransformer()),
                 ('model', RandomForestClassifier(n_estimators=100, criterion="entropy"))])

rf_model = pipe.fit(X_train, y_train)
rf_prediction = rf_model.predict(X_test)
print("Точность на обучении 80% данных: {}%".format(round(metrics.accuracy_score(y_test, rf_prediction)*100,2)))

rf_full_model = pipe.fit(_finish_clear_data['title'], _finish_clear_data['is_fake'])
rf_full_prediction = rf_full_model.predict(X_test)
print("Точность на обучении всеми данными: {}%".format(round(metrics.accuracy_score(y_test, rf_full_prediction)*100,2)))
```

***Матрица для 80% обученого случайного леса***

```python
cm = metrics.confusion_matrix(y_test, rf_prediction)
plot_confusion_matrix(cm, classes=['Fake', 'Real'])
```

***Матрица для полностью обученного случайного леса***

```python
cm = metrics.confusion_matrix(y_test, rf_full_prediction)
plot_confusion_matrix(cm, classes=['Fake', 'Real'])
```
